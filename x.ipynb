```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request

# load datset into data
data = pd.read_csv('california.csv')
data
```

```python
# check for null value
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
# train test data split into 20 80 ratio
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
#Comvine ytrain data after spliting
train_data = X_train.join(Y_train)
train_data
```

```python
train_data.hist(figsize =(15,8))
```

```python
# fixed the plot size 
plt.figure(figsize=(15,8))

# drop the ocean proximity bcz w'are doing corellation MATRIX of train_data
train_dataa = train_data.drop(['ocean_proximity'], axis=1)

sns.heatmap(train_dataa.corr(), annot=True, cmap='YlGnBu')
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)

train_data.hist(figsize=(15,8))
```

```python
# get ocean proximity into categorical feature.
ocean = pd.get_dummies(train_data.ocean_proximity)
ocean

```

```python
# join those categorical data into train data
train_data = train_data.join(ocean)

#drop ocean proximity from train rata bcz we already divided ocean proximity into categorical from so we don't need ocen proximity.
train_data = train_data.drop(['ocean_proximity'], axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
# creating Linear regression Model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train,Y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg =LinearRegression()

reg.fit(X_train_s,Y_train)

```

```python
test_data = X_test.join(Y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

ocean = pd.get_dummies(test_data.ocean_proximity)
test_data = test_data.join(ocean)

#
test_data = test_data.drop(['ocean_proximity'], axis=1)

test_data['bedrooms_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
X_test, Y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s,Y_test)
```

```python
# crwating Random Forest Classifier
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s,Y_train)
```

```python
forest.score(X_test_s,Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators" : [3,10,30],
    "max_features" : [8,12,20],
    "min_samples_split" : [2,4,6,8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring="neg_mean_square_error",
                           return_train_score=True)

grid_search.fit(X_train_s, Y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, Y_test)
```
